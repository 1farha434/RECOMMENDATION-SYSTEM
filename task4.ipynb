```python
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = {
    'User': ['U1','U1','U1','U2','U2','U2','U3','U3','U3','U4','U4'],
    'Movie': ['A','B','C','A','B','D','A','C','D','B','C'],
    'Rating': [5,4,5,4,5,3,5,4,4,2,5]
}

df = pd.DataFrame(ratings)

print(df)
```

```python
user_movie_matrix = df.pivot_table(
    index='User',
    columns='Movie',
    values='Rating'
)

user_movie_matrix = user_movie_matrix.fillna(0)

print(user_movie_matrix)
```

```python
similarity = cosine_similarity(user_movie_matrix)

similarity_df = pd.DataFrame(
    similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)

print(similarity_df)
```

```python
def recommend_movies(user):

    similar_users = similarity_df[user].sort_values(
        ascending=False
    )

    similar_users = similar_users.drop(user)

    most_similar_user = similar_users.index[0]

    watched_movies = user_movie_matrix.loc[user]

    recommended = []

    for movie in user_movie_matrix.columns:

        if watched_movies[movie] == 0:

            if user_movie_matrix.loc[
                most_similar_user,
                movie
            ] > 0:

                recommended.append(movie)

    return recommended
```

```python
user = 'U4'

recommendations = recommend_movies(user)

print("Recommended Movies:", recommendations)
```

```python
for movie in recommendations:
    print("Recommended:", movie)
```
